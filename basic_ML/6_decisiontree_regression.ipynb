```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset = pd.read_csv('/content/Position_Salaries.csv')

#split x,y 
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

#  DecisionTreeRegressor
1. split the data 
2. take mean of the data which is on the right side and plot line
3. calculate sum of residuals (actual point - point on the line) <br>
   use mean squared error (MSE)
4. shift the splitting point by n points on x axis
5. repeat from (2) for n times 
6. find the splitting point where error is minimum (this is splitting node)
7. data which is on the right side of final splitting point is new data
8. repeat from (1) 



```
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X,y)
```

```
y_pred = regressor.predict(np.array([[6.5]]))
```

```
print(y_pred)
```

```
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=10,random_state=48)
rf.fit(X,y)
```

```
print(rf.predict(np.array([[6.5]])))
```
