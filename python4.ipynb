# Dict comprehension

```

feet_length = {'a': 1.02, 'b': 2.5, 'c': 2.5}

feet_to_inch = 12
inch_length = {item: value*feet_to_inch for (item, value) in feet_length.items()}
print(inch_length)
```

```
check_length = {key: value for (key, value) in feet_length.items() if value > 2}
print(check_length)
```

```
check_length = {key: ('in_range' if value > 2 else 'small') for (key, value) in feet_length.items()}
print(check_length)
```

# Traceback


```
import traceback
#traceback method file handling 
'''
try executes the code 
if any exection control goes to "exception"
"fianlly" blocks executes irrespective of error or execution
'''
f = open("test.txt",'r')

#file handling without traceback
try:
  f.write("Hello!") 
  f.close()
  print('file closed')
except Exception as e: # throw exeception because file is not in read mode 
  print (str(e))
  traceback.print_exc()   
finally: 
  f.close() #automatically close the file
  print('file closed due to exception')

print('\n#########################################################################\n')

#file handling with traceback
try:
  f.write("Hello!") 
  f.close()
  print('file closed')
except Exception as e: # throw exeception because file is not in write mode 
  print (str(e))
  traceback.print_exc()
finally: 
  f.close() #automatically close the file
  print('file closed due to exception')
```
