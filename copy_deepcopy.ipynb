Normal copy function

```
'''
normal copy function points new variable to same location of source variable
'''
a = [1, 2, 3, 4, 5, 6]  
b = a
print (id(a) == id(b))
print(id(a[0]) == id(b[0]))   
a.append(8)                                                                                                                                            
print(a)
print(b)
```

Shallow copy

```
'''
shallow copy makes new object with same references
but after copy operation, does not bind to new data with new ref
'''
import copy

a = [1, 2, 3, 4, 5, 6]
b = copy.copy(a) #shallow copy

a[0] = 0 #make change in B as well
a.append(10) #no changes in B
print("A", a)
print("B", b)
```

deep copy

```
'''
deepcopy makes diff copy of object with new mem location 
'''
import copy
a = [1, 2, 3, 4, 5, 6]  
b = copy.deepcopy(a)                     
print(id(a) == id(b)) 
a[5] = 7
a.append(8)                                                                                                                                        
print(a)
print(b)
```

Profiling


```
#time to execute instruction
import time 
def sort_list():
  my_list = [1,5,4,7,8,9,55,44,87,94,65,48,12,11,10]
  my_list.sort()
  print(my_list)


start = time.time() #time before calling function 
sort_list() 

print((time.time() - start), 'sec') #diff bet current time and time before calling function
```

```
'''
1: number of function calls and the time run
2: Ordered by: standard name means the text string in the far right column was used to sort the output
                This could be changed by the sort parameter.

3: onwards contain the functions and sub functions called internally
ncalls : number of calls made
tottime: Total time by given function (sub-functions are excluded)
percall: Total time / No of calls (remainder is left out )
cumtime: includes time spent in this and all subfunctions
percall: following cumtime is calculated as the quotient of cumtime divided by primitive calls. 
        The primitive calls include all the calls that were not included through recursion.
'''
import cProfile 
def my_func():
  for x in range(1,100):
    print(x)

cProfile.run('my_func()') 
```
