```
import requests 
from bs4 import BeautifulSoup 
import json
headers = {'User-Agent': 'Mozilla/5.0'} 


UNITS = {
         "cup": ["cups", "cup", "c.", "c"],
         "fluid_ounce": ["fl. oz.", "fl oz", "fluid ounce", "fluid ounces"],
         "gallon": ["gal", "gal.", "gallon", "gallons"],
         "ounce": ["oz", "oz.", "ounce", "ounces"],
         "pint": ["pt", "pt.", "pint", "pints"],
         "pound": ["lb", "lb.", "pound", "pounds"],
         "quart": ["qt", "qt.", "qts", "qts.", "quart", "quarts"],
         "tablespoon": ["tbsp.", "tbsp", "T", "T.", "tablespoon", "tablespoons", "tbs.", "tbs"],
         "teaspoon": ["tsp.", "tsp", "t", "t.", "teaspoon", "teaspoons"],
         "gram": ["g", "g.", "gr", "gr.", "gram", "grams"],
         "kilogram": ["kg", "kg.", "kilogram", "kilograms"],
         "liter": ["l", "l.", "liter", "liters"],
         "milligram": ["mg", "mg.", "milligram", "milligrams"],
         "milliliter": ["ml", "ml.", "milliliter", "milliliters"],
         "pinch": ["pinch", "pinches"],
         "dash": ["dash", "dashes"],
         "touch": ["touch", "touches"],
         "handful": ["handful", "handfuls"],
         "stick": ["stick", "sticks"],
         "inch": ["inch", "inches"],
         "clove": ["cloves", "clove"],
         "can": ["cans", "can"], "large": ["large"],
         "small": ["small"],
         "scoop": ["scoop", "scoops"],
         "filets": ["filet", "filets"],
         "sprig": ["sprigs", "sprig"],
         "glass": ["glass", "glasses"],
         "bunch": ["bunch", "bunches"],
         "stalk": ["stalk"],
         "shot": ["shots", "shot"]
         }


liquids = ['buttermilk', 'cottage', 'cream', 'creamer', 'creamy', 'creme', 'ghee', 'half-and-half',
           'milk', 'yogurt', 'dahi', 'juice', 'syrup', 'essence', 'honey', 'sauce', 'butter', 'oil']

nuts = ['almonds', 'butternuts', 'candlenuts', 'cashews', 'chestnuts', 'hazelnuts', 'macadamia', 'nuts', 'almond',
        'butternut', 'candlenut', 'cashew', 'chestnut', 'hazelnut'
        'peanuts', 'pecans', 'pistachios', 'walnuts', 'nuts', 'pine',
        'macadamia', 'nut', 'filberts', 'pistachio', 'pecan', 'chestnut', 'peanut', 'walnut']

grains = ['chana', 'dal', 'daal', 'dry', 'barley', 'bran', 'cereal', 'corn', 'cornstarch', 'granola', 'hominy', 'kasha',
          'millet', 'oats', 'popcorn', 'pretzels', 'quinoa', 'rice',
          'tapioca', 'breadcrumbs', 'bulgur', 'farina', 'oatmeal', 'polenta',
          'sago', 'seitan', 'grains', 'taters', 'cornflakes', 'chickpea', 'lentil', 'daal', 'sugar', 'oat', 'oats',
          'jowar', 'bajra', "dry"]

pastas = ['farfalle', 'fettuccine', 'lasagnas', 'linguine', 'mac', 'macaroni', 'manicotti', 'noodles', 'pasta',
          'farfel', 'vermicelli', 'tagliatelle', 'cannelloni', 'penne', 'spagetti'] 

def get_cal(measure_name,nutri_name,quantity):
  '''
  calculates calories for 100gm of the given food 

  Args:
        measure_name(str): unit measuring name
        nutri_name(str): food name
        quantity(float): food quantity

  Returns:
          int: grams present in given unit
  '''
  if measure_name in UNITS['cup'] or measure_name in UNITS['glass']:
      temp_weight = 150

      for word in liquids:
        if word in str(nutri_name.lower()):
          temp_weight = 255
          break

      for word in nuts:
        if word in str(nutri_name.lower()):
          temp_weight = 150
          break

      for word in grains:
        if word in str(nutri_name.lower()):
          temp_weight = 200
          break

      for word in pastas:
        if word in str(nutri_name.lower()):
          temp_weight = 200
          break
          
      return temp_weight    



  elif measure_name in UNITS['tablespoon']:
      return (13.63)

  elif measure_name in UNITS['teaspoon']:
      return (4.3)

  elif measure_name in UNITS["kilogram"]:
      return (1000)

  elif measure_name in UNITS["liter"]:
      return (1000)

  elif measure_name in UNITS["milligram"]:
      return (0.001)

  elif measure_name in UNITS["milliliter"]:
      return (quantity)  

  elif measure_name in UNITS["gram"]:
      return quantity

  elif measure_name in UNITS["dash"]:
      return (0.50)

  elif measure_name in UNITS["touch"]:
      return (0.35)

  elif measure_name in UNITS["pinch"]:
      return (0.35)

  elif measure_name in UNITS["stick"]:
      return (113)

  elif measure_name in UNITS["inch"]:
      return (13)

  elif measure_name in UNITS["bunch"]:
      return (250)

  elif measure_name in UNITS["clove"]:
      return (5)

  elif measure_name in UNITS["scoop"] or measure_name in UNITS["handful"]:
      return (10)

  elif measure_name in UNITS["sprig"]:
      return (0.001)

  elif measure_name in UNITS["filets"]:
      return (91)

  elif measure_name in UNITS["gallon"]:
      return (3785.41)

  elif measure_name in UNITS["pound"]:
      return (453.92)

  elif measure_name in UNITS["fluid_ounce"] or measure_name in UNITS["ounce"]:
      return (28.35)

  elif measure_name in UNITS["pint"]:
      return (473.17)

  elif measure_name in UNITS["stalk"]:
      return (50)

  elif measure_name in UNITS["shot"]:
      return (30)

  else:
    return (0) 


def get_nutritional_values(food_keyword):
  '''
  Search for the nutritional_values of a given food on webmd.com site
  clean the crawled data

  Args:
        food_keyword(str): name of the food 

  Returns:
          nutrients_data(dict): dictionary of nutrients and it's values
                                present in the given food
  '''

  #encode keyword if it contains whitespace
  if " " in food_keyword:
    food_keyword = food_keyword.replace(' ', '%20')


  URL = "https://www.webmd.com/ffp/resourceservice.svc/searchtype/FoodSearch\
        /phrase/"+food_keyword+"/sortby/default/direction/asc/skip/0/take/50\
        /category/empty"
  
  web_page = requests.get(URL, headers=headers)
  soup_object = BeautifulSoup(web_page.content, 'lxml')
  #data is present inside <body> -> <p> -> "data" -> "items" 
  raw_data = json.loads(soup_object.body.p.string)
  raw_data = raw_data["data"]["items"][0] #actual data 

  #delete irrelevant values 
  delete_items_list = ["sname", "manunum","pn", "rf", "ver", "id","sc","mu"]
  for items in delete_items_list:
    if items in raw_data:
          del raw_data[items] #delete current item from dictionary 


  replace_names = ['name','quantity','measure_name','measure_abbreviation','total_protein(gm)',
                   'total_calories','total_calcium','total_carbohydrates(gm)',
                   'total_sugar(gm)','total_sat_fat(gm)','total_fat(gm)',
                   'total_dietary_fiber(gm)','total_trans_fat(gm)','omega3',
                   'vitamin_d','total_cholesterol (mg)','vitamin_a',
                   'total_sodium(mg)','vitamin_c','total_iron'
                   ]
  nutrients_data = dict(zip(replace_names, list(raw_data.values())))

  if "%20" in food_keyword:
    food_keyword = food_keyword.replace('%20', ' ')

  nutrients_data["name"] = str(food_keyword)

  #round off the values returned by the server
  for current_nutrient in nutrients_data:
    if isinstance(nutrients_data[current_nutrient], float):
      nutrients_data[current_nutrient] = round(nutrients_data[current_nutrient])         


  '''
  1)values returned from the server are in mg/mcg which needs to be converted to
    daily percentage of nutrients.

  2)For each nutrient a specific daily intake has been recommended
    according to that intake, nutrient's percentage is calculated
    

  3)for calories from fat use formula (total fat * 9)
    1gm of fat gives 9 calories
  '''
  nutrients_data['calories_from_fat'] = round(nutrients_data["total_fat(gm)"] * 9)
  nutrients_data['protein_percentage'] = round(100.0 * nutrients_data["total_protein(gm)"] / 20)
  nutrients_data['total_fat_percentage'] = round(100.0 * nutrients_data["total_fat(gm)"] / 60)
  nutrients_data['total_sat_fat_percentage'] = round(100.0 * nutrients_data["total_sat_fat(gm)"] / 22)
  nutrients_data['total_cholesterol_percentage'] = round(100.0 * nutrients_data["total_cholesterol (mg)"] / 300)
  nutrients_data['total_sodium_percentage'] = round(100.0 * nutrients_data["total_sodium(mg)"] / 6000)
  nutrients_data['total_carbohydrates_percentage'] = round(100.0 * nutrients_data["total_carbohydrates(gm)"] / 275)
  nutrients_data['total_dietary_fiber_percentage'] = round(100.0 * nutrients_data["total_dietary_fiber(gm)"] / 28)
  nutrients_data['total_vitamin_c_percentage'] = round(100.0 * (nutrients_data["vitamin_c"]) / 82)
  nutrients_data['total_vitamin_a_percentage'] = round(100.0 * (nutrients_data["vitamin_a"]) / 800)
  nutrients_data['total_calcium_percentage'] = round(100.0 * (nutrients_data["total_calcium"]) / 2500)
  nutrients_data['total_iron_percentage'] = round(100.0 * (nutrients_data["total_iron"]) / 21)
  nutrients_data['quantity'] = float(nutrients_data['quantity'])

  measure_name = (nutrients_data['measure_name']).lower()
  nutri_name = (nutrients_data['name']).lower()
  quantity = nutrients_data['quantity']
  calories = nutrients_data['total_calories']

  gms_unit = get_cal(measure_name,nutri_name,quantity)

  if gms_unit == 0:
    nutrients_data['cal_100gm'] = 0
  else:   
    #converting cal to 100gm cal
    calories_100gm = round((100* calories) / (gms_unit * quantity))
    nutrients_data['cal_100gm'] = calories_100gm
  return nutrients_data
```
