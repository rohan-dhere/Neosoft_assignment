# Iterator

```
class sqr():
  def __init__(self,start,end):
    self.start = start
    self.end = end

  def __iter__(self):
    return self

  def __next__(self):
    if self.start <= self.end:
      ans = self.start * self.start
      self.start = self.start + 1 
      return ans
    else:
      raise StopIteration
```

```
myvalues = sqr(1,10)
```

```
while True:
  try:
    print(next(myvalues))
  except StopIteration:
    break
```

# Generators

```
def cir(start,end):
  while start < end:    
    yield (3.14 * start * start)
    start = start+1
  raise StopIteration
```

```
for x in (cir(1,10)):
  print(x)
```

```
def nextSquare(): 
    i = 1;
    while True: 
        yield i*i                 
        i += 1      
for num in nextSquare(): 
    if num > 100: 
         break    
    print(num) 
```
