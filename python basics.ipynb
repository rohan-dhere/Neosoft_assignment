Undefined length argument 


```
def mydata(name,*data): # we can pass undefined length arguments using *variable
  print(name)
  for i in data:
    print(i)

mydata('Rohan',1234,987654321,'DS')    
```

Undefined length argument with key value pair


```
def mydata(name,**data): #we can pass undefined length  with key-value pair arguments using **variable
  print(name)
  for i,j in data.items(): # i is key and j is value 
    print(i,':-',j)

mydata('Rohan',eid=1234,mob=987654321,dept='DS')    
```

# types of methods
class method use @classmethod <br>
static method use @staticmethod <br>
statc has nothing to do with class variables

```
class Student:

    school = 'xyz'

    def __init__(self,m1,m2,m3):
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3

    def avg(self):
        return (self.m1 + self.m2 + self.m3)/3

    @classmethod
    def getSchool(cls):
        return cls.school

    @staticmethod
    def info():
        print("This is Student class")

s1 = Student(45,68,48)
s2 = Student(44,96,45)



print(s1.avg())
print(Student.getSchool())

Student.info()
```

#Inheritance

```
class A:
    def feature1(self):
        print("F1")

    def feature2(self):
        print("F2")

class B:
    def feature3(self):
        print("F3")

    def feature4(self):
        print("F4")

class C(A,B):
    def feature5(self):
        print("F5")


a1 = A()

a1.feature1()
a1.feature2()
b1 = B()
c1 = C()


c1.feature1()
c1.feature2()
b1.feature1()
```

super method to access superclass methods in subclass


```
class A:
  def __init__(self):
    print('In class A')
  
  def feature1(self):
    print("F1")

  def feature2(self):
      print("F2")

class B:
  def __init__(self):
    print('In class B')

  def feature3(self):
      print("F3")

  def feature4(self):
      print("F4")

class C(A,B):
  def __init__(self):
    super().__init__()
    print('In class C')
  def feature5(self):
    print("F5")


c1 = C()
c1.feature1()
c1.feature2()
```
