# Zip()

```
list1 = [
         1,2,3,
         4,5,6
         ]
list2 = [
         'a','b','c',
         'd','e','f'
         ]
```

```
new_list = zip(list1,list2) #zip two list
print(new_list) #print only object, because zip returns iterator
```

```
for x,y in new_list:
  print(x,y)
```

```
for x,y in zip(list1,list2):
  print(x,y)
```

# Multithreading

```
from threading import * #for multithreading
from time import sleep

class hello(Thread): #Thread should be super class for multithreading
  def run(self): #run is a method in "Thread" 
    for x in range(100):
      print('Hello')
      sleep(0.5) #just to slow printing & check output 


class hi(Thread):
  def run(self):
    for x in range(100):
      print('hi')
      sleep(0.5)

t1 = hello()
sleep(0.5)
t2 = hi()

t1.start() #to create new t1 thread 
t2.start() #to create new t2 thread


t1.join() #waiting for t1 to finish & join main thread 
t2.join() #waiting for t2 to finish & join main thread
print('END')
```

# Socket programming
executed on local machine

```
#server 

import socket

server = socket.socket() #network socket object 
print('Socket created')

server.bind(('127.0.0.1',5001))  #binding ip & port to create socket

server.listen() #listen incoming connections
print('Waiting for connections...')

while True:
    client_port, address  = server.accept() #accept from incoming connection
    print('Connected with',address, client_port)
    
    client_port.send(bytes('Welcome...','utf-8')) #converting msg to bytes utf-8 format & sending
    client_port.close() 


#client
import socket

client = socket.socket() #network socket object 
client.connect(('127.0.0.1',5001)) #conencting  to ip & port 
msg = input('Enter message ')
client.send(bytes(msg,'utf-8')) #sending msg to server
print(client.recv(1024).decode()) #decode byte msg sent by server
```
