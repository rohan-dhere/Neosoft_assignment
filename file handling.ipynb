try catch with file handling


```
#traceback method file handling 
'''
try executes the code 
if any exection control goes to "exception"
"fianlly" blocks executes irrespective of error or execution
'''
f = open("test.txt",'r')
try:
  f.write("Hello!") 
  f.close()
  print('file closed')
except Exception as e: # throw exeception because file is not in read mode 
  print (str(e))
finally: 
  f.close() #automatically close the file
  print('file closed due to exeption')
```

Json file handling

```
import json

details = {
    'name': 'A',
    'tech': ['Python','R','Numpy','Pandas','TF']
}

with open('test.json', 'w') as json_file:
    json.dump(details, json_file)
    #json dumps
    #json loads
```
