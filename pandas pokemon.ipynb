#Label Encoding

```
def label_en(x):
  '''
  takes one element from series 
  returns label for given element 
  names : dictionary of labels
  get() function returns value of given key
  '''

  names = {
  'Water': 0,
  'Normal': 1,
  'Grass' : 2,
  'Bug': 3,
  'Psychic': 4,
  'Fire': 5,
  'Electric': 6,
  'Rock': 7,
  'Ghost': 8,
  'Dragon': 9,
  'Ground': 10,
  'Dark': 11,
  'Poison': 12,
  'Fighting': 13,
  'Steel': 14,
  'Ice': 15,
  'Fairy': 16,
  'Flying': 17
  }
  return names.get(x) #get value of current key/element of series

def leb(type1):
  '''
  takes one pandas series
  returns series with changed values
  # apply function applies operation to 
  every element of series
  '''
  return type1.apply(label_en)
```

```
type1 = df['Type 1']
temp_type1 = leb(type1)

type2 = df['Type 2']
temp_type2 = leb(type2)
# note: "apply" for a series & "applymap" for dataframe
```

```
temp_df = temp_df.drop(['Type 1','Type 2'], axis = 1)
temp_df['Type 1'] = temp_type1.values
temp_df['Type 2'] = temp_type2.values
temp_df.head()
```
