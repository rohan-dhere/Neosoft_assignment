```
from multiprocessing import Process
import traceback 

def fun1():
  while True:
    print(1)

def fun2():
  while True:
    print(2)

'''
create object of "Process" class
target: the function to be executed by process
start:  starts a thread
'''

if __name__ == '__main__':
  try:
    process1 = Process(target = fun1) # function 1 to be executed by process 1
    process1.start() # start the process1 activity.
    process2 = Process(target = fun2) #function 2 to be executed by process 2
    process2.start() # start process2 activity.
    
    # wait for processes to exit 
    process1.join() 
    process2.join()


  except:
    print('Some error!')
    traceback.print_exc()
    if process1.is_alive(): # check if the process1 is still alive
      process1.terminate()  
      print('process 1 terminated due to error')
    
    if process2.is_alive(): # check if the process2 is still alive
      process2.terminate()
      print('process 2 terminated due to error')
      
```
